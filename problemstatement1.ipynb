Model Building and Experimentation
1.Identification of the  type of problem
The above problem is a classification problem as we need to find out the whether the given customer would churn in the future or not(2 categories).
2.Identify suitable technique
For this problem, we are going to use decision tree classification as it is a simple classification algorithm.

```python
Data2_ind=data.copy()
Data2_ind=Data2_ind.drop('Exited',axis=1)
Data2_y=data[['Exited']]
Data2_ind=Data2_ind.drop('Surname',axis=1)
Data2_ind=Data2_ind.drop('CustomerId',axis=1)
```

```python
#Categorical variables need to be dummified before sending to the model
Data2_final=pd.get_dummies(Data2_ind)
```

```python
print(Data2_final.shape)
```

```python
Data2_final.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Data2_final, Data2_y, test_size=0.2, random_state=123)
```

```python
print(Data2_final.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC=DecisionTreeClassifier(max_depth=3) # Defining the model
DTC.fit(X_train,y_train) # building model on train
```

```python
preds=DTC.predict(X_train) # predictions on train using predict function
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_train,preds))
```

```python
accuracy_score(y_train,preds)
```

```python
DTC.fit(X_test,y_test)
pred2=DTC.predict(X_test)
accuracy_score(y_test,pred2)
```

```python
print(confusion_matrix(y_test,pred2))
```
